```python
#imports
import os
import pandas as pd
import numpy as np
import random
import warnings
warnings.simplefilter('ignore')
pd.set_option('display.max_columns', 500)

random.seed(1)
```

```python
df_15 = pd.read_csv('raw/2015.csv')
df_15.shape
```

```python
df_15 = df_15.loc[:, (df_15.isnull().sum() / df_15.shape[0]) < 0.1]
df_15.shape
```

```python
df_15 = df_15.dropna()
df_15.shape
```

```python
df_15 = df_15.drop('IDATE', axis=1)
```

```python
df_1114 = [pd.read_csv(f'raw/20{i}.csv') for i in range(11, 15)]
```

```python
df_list = [df_15] + df_1114
df = pd.concat(df_list, join='inner')
df.shape
```

```python
df = df.loc[:, (df.isnull().sum() / df.shape[0]) < 0.1].dropna()
df.shape
```

```python
df['DIABETE3'] = df['DIABETE3'].replace({2:0, 3:0, 1:2, 4:1})
df = df[(df['DIABETE3'] != 7) & (df['DIABETE3'] != 9)]
df.groupby(['DIABETE3']).size()
```

```python
df.loc[:, 'DIABETE3'] = df['DIABETE3'].replace({1:0}).replace({2:1})
df.groupby(['DIABETE3']).size()
```

```python
df = df.rename(columns={'DIABETE3': 'Diabetes'})
```

```python
df.dtypes
```

```python
for col in df.columns:
    if df[col].dtype != float:
        # print(col, 'str')
        continue
    if df[col].apply(float.is_integer).all():
        df = df.astype({col: 'int'})
        # print(col, df[col].dtype)
    else:
        # print(col, df[col].dtype)
        continue
df = df.astype({col: 'float' for col in ['WEIGHT2', 'HEIGHT3', '_BMI5', 'WTKG3', '_STSTR']})
```

```python
df_int = df.loc[:, df.dtypes == 'int64']
df_int = df_int.drop(['SEQNO', '_PSU'], axis=1)
```

```python
df = df.drop(['SEQNO', '_PSU'], axis=1)
# drop DIABETE3 for IID
df = df.drop("Diabetes", axis=1)
df.shape
```
